```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import locale
```

```python
url = "https://news.microsoft.com/pt-BR/recent-news/"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36",
}
```

```python
response = requests.get(url, headers=headers)
response_content = response.content.decode('UTF-8')

# <Response [200]> -> requisição foi bem-sucedida 
```

```python
soup = BeautifulSoup(response_content, 'html.parser')
```

```python
soup
```

```python
# Exibe o html como texto
print(soup.get_text())
```

```python
news_links = [a['href'] for a in soup.find_all("a", {"class": ["f-post-link", "c-heading-6", "m-chevron", "x-hidden-focus"]})]
news_links
```

```python
# Extraindo o título
title = soup.find("a", class_="f-post-link")["ms.title"]


# Extraindo as indexações (tags)
article = soup.find("article")
tags = [tag.replace("tag-", "") for tag in article["class"] if tag.startswith("tag-")]


# Extraindo a data (simulando uma data padrão, pois não há no HTML fornecido)

# Exibindo os resultados
print("Título:", title)
print("Indexações:", tags)
```
